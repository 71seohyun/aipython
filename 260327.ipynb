```python
height1 = float(input('학생1의 키(cm)를 입력하세요: '))
height2 = float(input('학생2의 키(cm)를 입력하세요: '))

if height1 == height2:
  print('두 학생의 키는 같다.')
elif height1 > height2:
  print('첫 번째 학생이 더 크다.')
else:
  print('두 번째 학생이 더 크다.')
```

```python
#1개의 점수를 입력
#90점 이상 A, 80점 이상 B, 70점 이상 C, 나머지 D로 출력하시오.

score = int(input('점수를 입력하세요.: '))

if score >= 90:
  print('A')
elif score >= 80:
  print('B')
elif score >=70:
  print('C')
else:
  print('D')
```

```python
import random

dice1 = random.randint(1, 6)
dice2 = random.randint(1, 6)

print(f'주사위1: {dice1}')
print(f'주사위2: {dice2}')

if dice1 == dice2:
  print('두 주사위의 값이 같다.')
elif dice1 > dice2:
  print('주사위1의 값이 주사위2의 값보다 크다.')
else:
  print('주사위2의 값이 주사위1의 값보다')
```

```python
#윤년 판별

year = int(input('연도를 입력하시오.: '))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
  print('윤년')
else:
  print('평년')
```

```python
import random

player1 = random.randint(1,3)
player2 = random.randint(1,3)

if player1 == 1:
  print('플레이어1: 가위')
elif player1 == 2:
  print('플레이어1: 바위')
else:
  print('플레이어1: 보')

if player2 == 1:
  print('플레이어2: 가위')
elif player2 == 2:
  print('플레이어2: 바위')
else:
  print('플레이어2: 보')

if player1 == player2:
  print('무승부입니다.')

else:
  if player1 == 1:
    if player2 == 2:
      print('플레이어2의 승리입니다.')
    else:
      print('플레이어1의 승리입니다.')

  elif player1 == 2:
    if player2 == 1:
      print('플레이어1의 승리입니다.')
    else:
      print('플레이어2의 승리입니다.')

  else:
    if player2 == 1:
      print('플레이어2의 승리입니다.')
    else:
      print('플레이어1의 승리입니다.')
```
