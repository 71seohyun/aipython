```python
joseon_kings = [
    "태조", "정종", "태종", "세종", "문종", "단종", "세조", "예종", "성종",
    "연산군", "중종", "인종", "명종", "선조", "광해군", "인조", "효종", "현종",
    "숙종", "경종", "영조", "정조", "순조", "헌종", "철종", "고종", "순종"
]
```

```python
for i, name in enumerate(joseon_kings, start=1):
  print(f'{i}대 왕 {name}')
```

```python
num = int(input('몇 대 왕을 찾을까요?: '))
if 1<=num<=27:
  print(f'{num}대 왕은 {joseon_kings[num-1]}입니다.')
else:
  print(f'{num}대 왕은 없습니다.')
```

```python
name = input('어떤 왕을 찾을까요?: ')
if name in joseon_kings:
  print(f'{name}은 {joseon_kings.index(name)+1}대 왕입니다.')
else:
  print(f'{name}은 없습니다.')
```

```python
for i, name in enumerate(list(reversed(joseon_kings))):
  print(f'{len(joseon_kings)-i}대 왕 {name}')
```
