```python
#출력

cafe_menu = {
    "아메리카노(ICE)": 4500,
    "아메리카노(HOT)": 4000,
    "카페라떼(ICE)": 5200,
    "카페라떼(HOT)": 5200,
    "바닐라라떼(ICE)": 5800,
    "카라멜마키아토(ICE)": 6100,
    "카푸치노(HOT)": 5500,
    "콜드브루": 5900,
    "말차라떼(ICE)": 6500,
    "자몽에이드": 6000
}

for key, value in cafe_menu.items():
  print(f'{key}의 값은 {value}원입니다.')
```

```python
#전체메뉴 값 합산

sum = 0

for i in cafe_menu.values():
  sum+=i

print(f'총 가격은 {sum}원입니다.')
```

```python
#주문받기

count = {}
cafeSum = 0

print("ㅡㅡㅡ메뉴ㅡㅡㅡ")
for k,i in cafe_menu.items():
  print(f'{k} : {i}')
print('ㅡㅡㅡㅡㅡㅡㅡㅡ\n')
print('주문받겠습니다. 주문을 끝내시려면 종료라고 말씀해주세요.')


while True:
  n = input().strip()
  if n in cafe_menu.keys():
    if n in count.keys():
      count[n]+=1
    else:
      count[n]=1
    cafeSum += cafe_menu[n]
  elif n == '종료':
    break
  else:
    print('다시 말씀해주시겠습니까?')

if not count:
  print('안녕히가세요.')
else:
  print(f'총 주문 ',end='')
  text=''
  for k,i in count.items():
    text+=f'{k} {i}잔, '
  text = text.rstrip(', ')
  print(text, end='')
  print(f'으로 총 {cafeSum}원입니다.')


```
