```python
#분을 넣으면 시간과 분을 계산해주는 프로그램
time=int(input("분을 입력해주세요: "))
hour=time//60
min=time%60
print(f'{time}분은 {hour}시간 {min}분입니다.')
```

```python
#산지니의 코드

#파이썬 코드로 다음 문제를 해결해줘.
#먼저 '분을 입력해주세요: '라는 프롬프트문을 제시하고 사용자에게 분을 입력받으면, '()분은 ()시간 (~)분입니다.' 라고 출력하는 프로그램을 만들어야해.

# 분을 입력받아 시간과 분으로 변환해 출력하는 프로그램

minutes = int(input("분을 입력해주세요: "))

hours = minutes // 60
remain = minutes % 60

print(f"{minutes}분은 {hours}시간 {remain}분입니다.")
```

```python
# 사용자에게 정수를 입력받아 짝수/홀수를 판별하는 프로그램

num = int(input("숫자를 입력해주세요: "))

# 어떤 수를 2로 나눴을 때 나머지가 0이면 짝수, 1이면 홀수입니다.
if num % 2 == 0:
    print("짝수입니다.")
else:
    print("홀수입니다.")
```

```python
# 사용자에게 정수를 입력받아 양수/음수를 판별하는 프로그램

num = int(input("숫자를 입력해주세요: "))

if num > 0:
    print("양수입니다.")
elif num < 0:
    print("음수입니다.")
else:
    print("0입니다.")
```

```python
"""
음료  가격
커피  2000원
라떼  3000원
녹차  2500원
주스  2800원

변수
"""

coffeePrice = 2000
lattePrice = 3000
greenteaPrice = 2500
juicePrice = 2800

coffeeCount = int(input('커피 주문수: '))
latteCount = int(input('라떼 주문수: '))
greenteaCount = int(input('녹차 주문수: '))
juiceCount = int(input('주스 주문수: '))

coffeePay = coffeeCount * coffeePrice
lattePay = latteCount * lattePrice
greenteaPay = greenteaCount * greenteaPrice
juicePay = juiceCount * juicePrice

total = coffeePay + lattePay + greenteaPay + juicePay

print(f'주문금액: {total} 원')
```
