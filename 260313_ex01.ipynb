```python
# 사칙연산(+, -, *, /)을 하는 아주 간단한 프로그램
# 변수(값을 담는 상자), 연산자(계산 기호), 출력문(print)만 사용

a = 12          # 첫 번째 숫자
b = 5           # 두 번째 숫자

add_result = a + b     # 더하기
sub_result = a - b     # 빼기
mul_result = a * b     # 곱하기
div_result = a / b     # 나누기 (파이썬에서는 / 결과가 소수일 수 있어요)

print("a =", a)                      # 변수 값 출력
print("b =", b)                      # 변수 값 출력
print("a + b =", add_result)         # 더하기 결과 출력
print("a - b =", sub_result)         # 빼기 결과 출력
print("a * b =", mul_result)         # 곱하기 결과 출력
print("a / b =", div_result)         # 나누기 결과 출력
```

```python
# 사칙연산(+, -, *, /)을 하는 아주 간단한 프로그램
# + 입력(input)을 받아서 a, b를 정하도록 추가한 버전

a = int(input("첫 번째 숫자 a를 입력하세요: "))   # 입력받은 문자열을 정수로 바꿔서 a에 저장
b = int(input("두 번째 숫자 b를 입력하세요: "))   # 입력받은 문자열을 정수로 바꿔서 b에 저장

add_result = a + b     # 더하기
sub_result = a - b     # 빼기
mul_result = a * b     # 곱하기
div_result = a / b     # 나누기 (0으로 나누면 에러가 나요)

print("a =", a)                      # 변수 값 출력
print("b =", b)                      # 변수 값 출력
print("a + b =", add_result)         # 더하기 결과 출력
print("a - b =", sub_result)         # 빼기 결과 출력
print("a * b =", mul_result)         # 곱하기 결과 출력
print("a / b =", div_result)         # 나누기 결과 출력
```
